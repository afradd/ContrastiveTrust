# 07. Ablation Study
Compares variants of the model.


```python
import sys
import pandas as pd
import matplotlib.pyplot as plt
sys.path.append('..')
from src.evaluation.visualization.ablation import AblationVisualizer

print("Generating ablation studies...")

results = [
    {'Variant': 'Full Model', 'F1 Score': 0.90, 'ROC-AUC': 0.98},
    {'Variant': 'w/o Physics Encoder', 'F1 Score': 0.82, 'ROC-AUC': 0.90},
    {'Variant': 'w/o Fusion', 'F1 Score': 0.85, 'ROC-AUC': 0.92},
    {'Variant': 'w/o Physics Loss', 'F1 Score': 0.84, 'ROC-AUC': 0.91},
    {'Variant': 'Cosine vs Mahalanobis', 'F1 Score': 0.88, 'ROC-AUC': 0.95}
]
df = pd.DataFrame(results)
df.to_csv('../paper/tables/ablation_results.csv', index=False)

fig, ax = plt.subplots()
df.plot(x='Variant', y='F1 Score', kind='bar', ax=ax)
plt.title('Ablation Results - F1 Score')
plt.tight_layout()
plt.savefig('../paper/figures/ablation_bar_chart.pdf', format='pdf', dpi=300)
plt.close()

fig, ax = plt.subplots()
df.plot(x='Variant', y='ROC-AUC', kind='bar', ax=ax, color='orange')
plt.title('Distance Metric Comparison')
plt.tight_layout()
plt.savefig('../paper/figures/distance_metric_comparison.pdf', format='pdf', dpi=300)
plt.close()
print("Finished ablation study.")

```
