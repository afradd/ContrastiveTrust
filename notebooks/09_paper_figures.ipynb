# 09. Paper Figures
Exports high-quality IEEE figures (PNG, PDF, SVG).


```python
import os
import sys
import matplotlib.pyplot as plt
import numpy as np

# Apply IEEE style if scienceplots is available, else standard fallback
try:
    import scienceplots
    plt.style.use(['science', 'ieee'])
    print("Using IEEE scienceplots style.")
except ImportError:
    plt.rcParams.update({
        'font.family': 'serif',
        'font.serif': ['Times New Roman'],
        'font.size': 8,
        'axes.labelsize': 8,
        'axes.titlesize': 8,
        'legend.fontsize': 6,
        'xtick.labelsize': 6,
        'ytick.labelsize': 6,
        'figure.dpi': 300
    })
    print("Using fallback IEEE styles.")

# Dummy plot to export in all formats
fig, ax = plt.subplots(figsize=(3.5, 2.5))
x = np.linspace(0, 10, 100)
ax.plot(x, np.sin(x), label='Signal A')
ax.plot(x, np.cos(x), label='Signal B', linestyle='--')
ax.set_xlabel('Time (s)')
ax.set_ylabel('Amplitude')
ax.legend()

print("Exporting final figures...")
base_path = '../paper/figures/final_overview'
plt.savefig(base_path + '.png', dpi=300, format='png', bbox_inches='tight')
plt.savefig(base_path + '_highres.png', dpi=600, format='png', bbox_inches='tight')
plt.savefig(base_path + '.pdf', format='pdf', bbox_inches='tight')
plt.savefig(base_path + '.svg', format='svg', bbox_inches='tight')
plt.close()
print("Finished paper figures export.")

```
