# 10. Paper Tables
Compiles tables into CSV, Excel, and LaTeX formats.


```python
import os
import pandas as pd

print("Compiling final paper tables...")

# Example: Dataset statistics
data = {
    'Dataset': ['SWaT'],
    'Sensors': [51],
    'Normal Windows': [40000],
    'Attack Windows': [5000]
}
df_stats = pd.DataFrame(data)
df_stats.to_csv('../paper/tables/dataset_statistics.csv', index=False)
try:
    df_stats.to_excel('../paper/tables/dataset_statistics.xlsx', index=False)
except ModuleNotFoundError:
    print("openpyxl not installed, skipping Excel export.")

# Compatibility for pandas versions regarding to_latex
with open('../paper/tables/dataset_statistics.tex', 'w') as f:
    if hasattr(df_stats.style, 'to_latex'):
        f.write(df_stats.style.to_latex(caption='Dataset Statistics', label='tab:dataset_stats'))
    else:
        f.write(df_stats.to_latex(index=False, caption='Dataset Statistics', label='tab:dataset_stats'))

try:
    df_metrics = pd.read_csv('../paper/tables/metrics.csv')
    with open('../paper/tables/metrics.tex', 'w') as f:
        if hasattr(df_metrics.style, 'to_latex'):
            f.write(df_metrics.style.to_latex(caption='Evaluation Metrics', label='tab:metrics'))
        else:
            f.write(df_metrics.to_latex(index=False, caption='Evaluation Metrics', label='tab:metrics'))
except FileNotFoundError:
    print("metrics.csv not found yet.")
print("Finished paper tables export.")

```
