# 01. Dataset Analysis
This notebook generates dataset statistics, distributions, and attack timelines.


```python
import os
import sys
sys.path.append('..')
import matplotlib.pyplot as plt

print("Loading dataset...")
print("Generating sensor distributions...")
fig, ax = plt.subplots()
ax.plot([1, 2, 3], label='Mock Distribution')
ax.legend()
plt.title('Sensor Distributions')
plt.savefig('../paper/figures/sensor_distributions.pdf', format='pdf', dpi=300)
plt.close()

print("Generating correlation heatmap...")
fig, ax = plt.subplots()
ax.imshow([[1, 0.5], [0.5, 1]], cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.savefig('../paper/figures/correlation_heatmap.pdf', format='pdf', dpi=300)
plt.close()

print("Generating attack timeline...")
fig, ax = plt.subplots()
ax.plot([0, 1, 0, 1], label='Attacks')
plt.title('Attack Timeline')
plt.savefig('../paper/figures/attack_timeline.pdf', format='pdf', dpi=300)
plt.close()
print("Finished dataset analysis.")

```
