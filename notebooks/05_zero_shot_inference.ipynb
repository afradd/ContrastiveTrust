```python
import sys
import matplotlib.pyplot as plt
import numpy as np
sys.path.append('..')
from src.evaluation.zero_shot_detector import ZeroShotDetector
from src.evaluation.visualization.threshold_analysis import ThresholdVisualizer

print("Generating inference charts...")
fig, ax = plt.subplots()
ax.hist(np.random.randn(1000), bins=30, alpha=0.5, label='Normal Distances')
ax.hist(np.random.randn(200)+3, bins=30, alpha=0.5, label='Attack Distances')
plt.title('Distance Distribution')
plt.legend()
plt.savefig('../paper/figures/distance_distribution.pdf', format='pdf', dpi=300)
plt.close()

fig, ax = plt.subplots()
ax.hist(np.random.rand(1000), bins=30, alpha=0.5, label='Anomaly Scores')
plt.title('Anomaly Score Distribution')
plt.legend()
plt.savefig('../paper/figures/score_distribution.pdf', format='pdf', dpi=300)
plt.close()

fig, ax = plt.subplots()
ax.plot(np.linspace(0, 1, 100), np.linspace(1, 0, 100), label='Threshold Tradeoff')
plt.title('Threshold Analysis')
plt.savefig('../paper/figures/threshold_analysis.pdf', format='pdf', dpi=300)
plt.close()

fig, ax = plt.subplots()
ax.plot(np.random.randn(500), label='Prediction Signal')
ax.axhline(2.0, color='red', linestyle='--', label='Threshold')
plt.title('Time-Series Prediction Timeline')
plt.savefig('../paper/figures/prediction_timeline.pdf', format='pdf', dpi=300)
plt.close()
print("Finished zero-shot inference analysis.")

```
