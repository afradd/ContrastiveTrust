# 02. Model Architecture
Generates parameter counts and model summaries.


```python
import os
import sys
sys.path.append('..')
from src.models.encoder import DualStreamEncoder, EncoderConfig
import pandas as pd

# Load architecture
config = EncoderConfig()
model = DualStreamEncoder(config)

total_params = sum(p.numel() for p in model.parameters())
trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)

print(f"Total Parameters: {total_params}")
print(f"Trainable Parameters: {trainable_params}")

# Generate table
summary_df = pd.DataFrame([{
    'Module': name,
    'Parameters': sum(p.numel() for p in module.parameters())
} for name, module in model.named_children()])

summary_df.to_csv('../paper/tables/model_summary.csv', index=False)
summary_df.to_csv('../paper/tables/parameter_breakdown.csv', index=False)
print("Finished model architecture analysis.")

```
