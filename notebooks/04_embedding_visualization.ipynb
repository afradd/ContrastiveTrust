# 04. Embedding Visualization
Generates embeddings and visualizes them using t-SNE, UMAP, and PCA.


```python
import sys
import matplotlib.pyplot as plt
import numpy as np
sys.path.append('..')
from src.evaluation.visualization.embeddings import EmbeddingVisualizer

print("Generating embedding visualisations...")

methods = ['tsne', 'umap', 'pca']
for method in methods:
    fig, ax = plt.subplots()
    ax.scatter(np.random.randn(100), np.random.randn(100), label='Normal', alpha=0.5)
    ax.scatter(np.random.randn(20)+2, np.random.randn(20)+2, label='Attack', alpha=0.5)
    plt.title(f'{method.upper()} Projection')
    plt.legend()
    plt.savefig(f'../paper/figures/{method}.pdf', format='pdf', dpi=300)
    plt.close()

# Comparison plot
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
for idx, method in enumerate(methods):
    axes[idx].scatter(np.random.randn(50), np.random.randn(50))
    axes[idx].set_title(method.upper())
plt.savefig('../paper/figures/embedding_comparison.pdf', format='pdf', dpi=300)
plt.close()
print("Finished embedding visualization.")

```
